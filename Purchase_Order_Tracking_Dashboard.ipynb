2. Imports & Utility Functions

```python
import pandas as pd
import numpy as np
from datetime import datetime
from IPython.display import display, HTML

# Utility: styled DataFrame display
def styled_df(df, caption=None):
    styler = df.style.set_table_styles([
        {'selector': 'th', 'props': [('background-color', '#333'), ('color', 'white')]},
        {'selector': 'caption', 'props': [('caption-side', 'bottom'), ('font-size', '1.1em')]}
    ])
    if caption:
        styler = styler.set_caption(caption)
    return styler.background_gradient(cmap='Blues', subset=df.select_dtypes(include=[np.number]).columns)
```

3. Data Loading Module

```python
# 3.1 Filenames
FILES = {
    'customers': 'olist_customers_dataset.csv',
    'geolocation': 'olist_geolocation_dataset.csv',
    'order_items': 'olist_order_items_dataset.csv',
    'payments': 'olist_order_payments_dataset.csv',
    'reviews': 'olist_order_reviews_dataset.csv',
    'orders': 'olist_orders_dataset.csv',
    'products': 'olist_products_dataset.csv',
    'sellers': 'olist_sellers_dataset.csv',
    'category_translation': 'product_category_name_translation.csv'
}

# 3.2 Load into DataFrames
dfs = {name: pd.read_csv(fname, parse_dates=True, infer_datetime_format=True)
       for name, fname in FILES.items()}

# 3.3 Preview schemas
for name, df in dfs.items():
    print(f"--- {name} ---")
    print(df.info(), '\n')
```

4. Data Cleaning & Date Parsing

```python
# 4.1 Standardize datetime columns in `orders`
date_cols = ['order_purchase_timestamp', 'order_approved_at',
             'order_delivered_carrier_date', 'order_delivered_customer_date',
             'order_estimated_delivery_date']
for col in date_cols:
    dfs['orders'][col] = pd.to_datetime(dfs['orders'][col], errors='coerce')

# 4.2 Fill missing numeric with zeros where appropriate
dfs['order_items']['shipping_limit_date'] = pd.to_datetime(
    dfs['order_items']['shipping_limit_date'], errors='coerce')
```

5. Feature Engineering & Joins

```python
# 5.1 Merge product categories
products = dfs['products'].merge(
    dfs['category_translation'], how='left', on='product_category_name'
)

# 5.2 Build master order-item table (without 'brand' column)
master = (
    dfs['orders']
    .merge(dfs['order_items'], on='order_id', how='left')
    .merge(dfs['payments'], on='order_id', how='left')
    .merge(products[['product_id','product_category_name_english']], on='product_id', how='left')
    .merge(dfs['sellers'], on='seller_id', how='left')
    .merge(dfs['customers'], on='customer_id', how='left')
)

# 5.3 Compute delivery delay (days)
master['delivery_delay'] = (
    master['order_delivered_customer_date'] - master['order_purchase_timestamp']
).dt.days

# 5.4 Extract year-month
master['order_month'] = master['order_purchase_timestamp'].dt.to_period('M').astype(str)

# Display basic info about the master table
print(f"Master table shape: {master.shape}")
print(f"Columns: {list(master.columns)}")
```

6. KPI Calculations Module

```python
# 6.1 Overall metrics
total_orders = master['order_id'].nunique()
avg_cycle_time = master['delivery_delay'].mean()
on_time_rate = (master['delivery_delay'] <= 0).mean()
total_revenue = master['payment_value'].sum()

html_summary = f"""
<div style="padding:12px;border:2px solid #2E86C1;border-radius:8px;background:#F0F8FF;">
  <h2 style="color:#2E86C1;">üìä Key Procurement Metrics</h2>
  <ul style="font-size:1.1em;">
    <li><b>Total Orders:</b> {total_orders}</li>
    <li><b>Total Spend:</b> ${total_revenue:,.2f}</li>
    <li><b>Avg. Cycle Time:</b> {avg_cycle_time:.1f} days</li>
    <li><b>On-Time Delivery Rate:</b> {on_time_rate:.1%}</li>
  </ul>
</div>
"""
display(HTML(html_summary))
```

19. Predictive Model A: Demand Forecasting (Monthly Orders)

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# Prepare time-series features (month index)
monthly['month_index'] = range(len(monthly))
X = monthly[['month_index']]
y = monthly['orders_count']

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, shuffle=False
)

model_demand = LinearRegression()
model_demand.fit(X_train, y_train)

# Forecast next 6 months
future_idx = pd.DataFrame({'month_index': range(len(monthly), len(monthly)+6)})
forecast = model_demand.predict(future_idx)

# CORRECTED: Generate future month labels properly
last_month_period = pd.Period(monthly['order_month'].iloc[-1], freq='M')
future_months = [(last_month_period + i + 1).strftime('%Y-%m') for i in range(6)]

# Visualize forecast
fig = go.Figure()
fig.add_trace(go.Scatter(x=monthly['order_month'], y=y, name='Actual'))
fig.add_trace(go.Scatter(x=future_months, y=forecast, name='Forecast', line=dict(dash='dash')))
fig.update_layout(title="Demand Forecast: Monthly Orders", xaxis_title="Month", yaxis_title="Orders")
fig.show()

# Print model performance
print(f"Training R¬≤ Score: {model_demand.score(X_train, y_train):.4f}")
print(f"Testing R¬≤ Score: {model_demand.score(X_test, y_test):.4f}")
print(f"Forecast for next 6 months: {forecast}")
```

20. Predictive Model B: Delivery-Delay Classification

```python
# First, let's check what product-related columns are available
print("Available columns in master DataFrame:")
product_cols = [col for col in master.columns if 'product' in col.lower()]
print("Product columns:", product_cols)

weight_cols = [col for col in master.columns if 'weight' in col.lower()]
print("Weight columns:", weight_cols)

numeric_cols = master.select_dtypes(include=[np.number]).columns.tolist()
print("Available numeric columns:", numeric_cols)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split
import numpy as np

# Prepare data: on-time (0) vs. delayed (1)
master['delay_flag'] = (master['delivery_delay'] > 0).astype(int)

# Use available numeric features (excluding IDs and target variables)
features = ['payment_value', 'freight_value', 'price', 'payment_installments', 'seller_zip_code_prefix']
df_ml = master.dropna(subset=features + ['delay_flag'])

print(f"Dataset shape after dropping NAs: {df_ml.shape}")
print(f"Features used: {features}")

X = df_ml[features]
y = df_ml['delay_flag']

# Check class distribution
print(f"\nClass Distribution:")
print(f"On-time deliveries (0): {(y == 0).sum()} ({(y == 0).mean():.1%})")
print(f"Delayed deliveries (1): {(y == 1).sum()} ({(y == 1).mean():.1%})")

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Train Random Forest model
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)

# Make predictions
y_pred = clf.predict(X_test)
y_pred_proba = clf.predict_proba(X_test)

# Display results
print("\n" + "="*50)
print("ü§ñ DELIVERY DELAY PREDICTION MODEL RESULTS")
print("="*50)
print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# Feature importance
feature_importance = pd.DataFrame({
    'Feature': features,
    'Importance': clf.feature_importances_
}).sort_values('Importance', ascending=False)

print("\nFeature Importance:")
for _, row in feature_importance.iterrows():
    print(f"{row['Feature']}: {row['Importance']:.4f}")
```
